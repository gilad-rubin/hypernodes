### Main Visualization Test

```python
from hypernodes import Pipeline, node


# 1. Define simple nodes
@node
def add(x: int, y: int) -> int:
    return x + y


@node
def multiply(a: int, b: int) -> int:
    return a * b


# 2. Create a pipeline
pipe = Pipeline([add, multiply])

print("Pipeline created successfully.")
```

```python
# 3. Visualize interactively
# This should render the widget below.
pipe.visualize(interactive=True)
```
