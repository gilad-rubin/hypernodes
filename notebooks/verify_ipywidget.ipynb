## Minimal ipywidget viz debug
Run cells top-down. If the widget stays blank, note any error text shown in the cell output (the widget now surfaces JS errors inline).

```python
from hypernodes import Pipeline, node

print("hypernodes module:", Pipeline.__module__)
```

```python
@node(output_name="y")
def add1(x: int):
    return x + 1


p = Pipeline(nodes=[add1], name="Demo")
w = p.visualize(engine="ipywidget")
print("Widget type:", type(w))
print("Iframe length:", len(w.value))
w
```

```python
# Decode the iframe HTML so we can see fallback/error text if rendering fails
import base64

html = base64.b64decode(w.value.split("base64,", 1)[1]).decode(
    "utf-8", errors="replace"
)
print(html[:800])
```
